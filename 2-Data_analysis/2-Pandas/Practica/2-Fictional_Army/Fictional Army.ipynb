# Fictional Army - Filtering and Sorting

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, None, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, None, 3, None, 2, None, 2, None, None, 1, 2, None],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

### Import libraries

### Step 1. This is the data given as a dictionary. Create a dataframe and assign it to a variable called army. 

```python
import numpy as np
import pandas as pd
```

```python
army = pd.DataFrame(raw_data)
```

### Step 2. Check missing values

```python
# sumo los nulos del dataframe
nans = army.isnull().sum()
print(type(nans))
print(nans)

# filtro aquellos que tengan algún nulo
nans = nans[nans > 0]
print('\n', nans)
```

### Step 3. Drop features, if there are more than 30% missing values.

```python
missing_30 = len(army) * 0.7
army.dropna(axis = 1, thresh = missing_30) # para que sobreescriba el dataframe, podríamos utilizar inplace=True
```

```python
# calculate the percentage of missing values for each column
missing_percentages = (army.isnull().sum() / len(army)) * 100
print(missing_percentages)

# check those columns with more than 30% of missing values
missing_percentage_30 = missing_percentages[missing_percentages > 30].index
print(missing_percentage_30)

# remove columns
cleaned_army = army.drop(columns = missing_percentage_30)
cleaned_army
```

### Step 4. Fill missing values with the mean of their regiment.

```python
cleaned_army[cleaned_army.isnull().any(axis=1)]
```

```python
# identificar la localización del NaN
# identificar su regimiento
cleaned_army[cleaned_army.isnull().any(axis=1)]['regiment']

# obtener la media de su regimiento
scouts_death_mean = cleaned_army.loc[cleaned_army['regiment'] == 'Scouts', 'deaths'].mean()
print(scouts_death_mean)

# aplico la máscara para que sólo en caso de que sea Scouts y deaths sustituya el dato
cleaned_army.loc[(cleaned_army['regiment'] == 'Scouts') & (cleaned_army['deaths'].isnull()), 'deaths'] = scouts_death_mean
```
