```python
import matplotlib.pyplot as plt

# Datos graciosos
etiquetas = [
    "Pepperoni (pero más pequeño)", 
    "Champiñones (nos conspiraron)", 
    "Queso extra (porque sí)", 
    "Anchoas (pero con bigote)"
]
tamaños = [30, 15, 45, 10]
explosion = (0, 0, 0.1, 0)  # Hacer destacar la porción de queso
colores = ["#FF4500", "#228B22", "#FFFF00", "#00008B"]

# Crear gráfico
plt.figure(figsize=(8, 6))
plt.pie(tamaños, 
        explode=explosion,
        labels=etiquetas,
        colors=colores,
        autopct='%1.1f%%',
        startangle=140,
        wedgeprops={'edgecolor': 'black'})

plt.title("Análisis científico 2024 de pizza", fontdict={'fontsize': 16, 'fontweight': 'bold'}, pad=20)
plt.ylabel("Informe confidencial del laboratorio de la cocina", fontsize=12, style='italic')
plt.gcf().set_facecolor('#f7f7f7')
plt.show()
```
