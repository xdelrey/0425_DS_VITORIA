```python
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import csv
import time
```

```python
url = "https://www.imdb.com/chart/top/"
```

```python
# /html/body/div[1]/main/div/div[2]/section/div/div[1]/div/ul/li[n]/div/div/div/div/div[1]/div/a
'''
import re

all_a = soup.find_all('a', class_=re.compile(r'\bipc-title-link-wrapper\b'))
print(len(all_a))

for x in all_a:
    print(x.get_text(strip=True))
'''
```

```python
from fake_useragent import UserAgent
ua = UserAgent()
headers = {'User-Agent': ua.random}
response = requests.get(url, headers=headers)
response
```

```python
html = response.text
print(html)
soup = bs(html, 'html.parser')
print(soup)
```

```python
film = {'Título:': [],
        'Año:': [],
        'Duración:': [],
        'Posición:': [],
        'Rating:': []}

for x in soup.find_all('a', class_='ipc-title-link-wrapper'):
    film['Posición:'].append(x.get_text().split('. ')[0])
    film['Título:'].append(x.get_text().split('. ')[1])

for x in soup.find_all('div', class_='sc-4b408797-7 fUdAcX cli-title-metadata'):
    film['Año:'].append(x.getText(',').split(',')[0])
    film['Duración:'].append(x.getText(',').split(',')[1])
    film['Rating:'].append(x.getText(',').split(',')[2])

film
```

```python
peli = pd.DataFrame(film)
peli
```

```python
peli.to_csv('top25_películas.csv', index_label='ID')
```

```python
# 3. Iterar y extraer
for idx, element in enumerate(data['itemListElement'], start=1):
    movie = element['item']
    title    = movie.get('alternateName') or movie.get('name')
    duration = movie.get('duration')  # ej. "PT2H22M"
    rating   = movie.get('aggregateRating', {}).get('ratingValue')
    print(f"{idx:3d}. {title} — {duration[2:].lower()}, Rating: {rating}")
```
